```python
a = 7

if a < 4:
    print(f"{a} is less than 4")
else:
    print (f"{a} is bigger than 4")
    
a < 4    
```

```python
# nested
a = 8

if a < 5:
    print("a < 5")
else:
    if a < 10:
        print("5 <= a < 10")
    else:
        print("a >= 10")
```

```python
# elif
a = 13

if a < 5:
    print('a < 5')
elif a < 10:
    print("5 <= a < 10")
elif a < 15:
    print("10 <= a < 15")
elif a < 20:
    print("15 <= a < 20")
else:
    print('a > 20')
```

```python
Conditional Expression ( Ternary Operator )  : X if (condition is true) else Y  
```

```python
a = 25 

if a < 5 :
    b = 'a < 5'
else:
    b = 'a >= 5'

print(b)    
```

```python
b = 'a < 5' if a < 5 else 'a >= 5'

print (b)
```

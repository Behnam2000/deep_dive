```python
min_length = 2

while True :
    name = input("Please enter your name")
    if len(name) >= min_length and name.isprintable() and name.isalpha():
        break

print("Hello , {0}".format(name))
```

2- "continue" :  do not allow us to continue on to the print statement but it still allow the loop to continue :

```python
a = 0 

while a < 10:
    a += 1
    if a % 2 == 0:
        continue  
    print(a)        
```

- The "else" clause
    - The the while loop also has an **else** clause that can be attached to it which executes if and only if
       the while loop terminated normally i.e. without a break statement. 

the "else" clause of a while loop will execute if and only if the while loop did not encounter a "break" statement

```python
# without usig 'else'

l = [1 , 2 , 3]
val = 10

found = False       # Flag 
idx = 0
while idx < len(l):
    if l[idx] == val:
        found = True
        break
    idx += 1

if not found:
    l.append(val)

print(l)
```

```python
# with using 'else'

l = [1 , 2 , 3]
val = 10
idx = 0

while idx < len(l):
    if l[idx] == val:
        break
    idx += 1
else:
    l.append(val)

print(l)

```

#### Break , Continue and the Try Steatement

- And then the **continue** and **break** statements can be used inside a try statement inside a loop, and the **finally** will always still      execute before either breaking or continuing from that point forward.

- try ...except...finally : the finally clause always runs even when there's an exception.

```python
a = 10 
b = 0

try:
    a/b
    
except ZeroDivisionError:
    print('Error : division by 0')
    
finally:
    print('this always executes')
    
```

- if you are inside a statement and you have a **continue** , and if you have a **finally** , the **finally** will execute :   

```python
a = 0
b = 2

while a < 4:
    print ("--------------")
    a += 1
    b -= 1

    try:
        a / b
    except ZeroDivisionError:
        print('{0} , {1} - division by 0'.format(a , b))
        continue
    finally:
        print("{0} , {1} - always executes".format(a , b))

    print("{0} , {1} - main loop".format(a , b))
```

- and this works with **break** as well ,  not just **continue**
</br>
  we're going to break as soon as we hit the first error.
  
  and we can also combine this with the **else** clause.
  
  Q : when will the **else** executes? A : if the **break** was never encountered.

```python
a = 0
b = 10

while a < 4:
    print ("--------------")
    a += 1
    b -= 1

    try:
        a / b
    except ZeroDivisionError:
        print('{0} , {1} - division by 0'.format(a , b))
        break
    finally:
        print("{0} , {1} - always executes".format(a , b))

    print("{0} , {1} - main loop".format(a , b))

else:
    print("Code is executed without a 'ZeroDivisionError'.")
```

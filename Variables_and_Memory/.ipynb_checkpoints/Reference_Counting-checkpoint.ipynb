# Reference Counting

### sys module

```python
import sys
```

```python
a = [1, 2, 3]
```

```python
id(a)
```

- We created an object in memory and **a** is points to that memory address , **a** is the pointer

```python
sys.getrefcount(a) # it's going to point to the same object in memory
```

### ctypes module

```python
import ctypes
```

```python
def ref_count(address: int):
    return ctypes.c_long.from_address(address).value
```

```python
ref_count(id(a))
```

```python
ref_count(1746788169024)
```

### ! Question
If we get **sys.getrefcount(a)** and it gave us **2** because we passed **a** into this function , so why **ref_count(id(a))** gave us **1**?  

### ! Answer
- What is happening is id of **a** is getting evaluated first.
- So then the id finishes running and returns the memory address , and by the time we call **ref_count** , id has finished running and it has released it's pointer to that memory address

```python
b = a
```

```python
id(a)
```

```python
ref_count(id(a))
```

```python
c = a
ref_count(id(a))
```

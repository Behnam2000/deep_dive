# Variables and Memory References

```python
my_var = 10
```

**10** is stored somewhere in memory and **my_var** is now a reference to that object. 
    - so now Python will look at **my_var** and what is the memory address that it is referencing 
    it'll find that memory address 
    it'll go to the memory
    retrieve the data from memory
    and then brought it back so we can display it

```python
print(my_var)
```

```python
print(id(my_var))
```

```python
print(hex(id(my_var)))
```

```python
greet = 'hello'
```

```python
print(greet)
```

```python
print(id(greet))
```

```python
print(hex(id(greet)))
```

### Bottom line:
- Variables are just memory addresses.
- They not actually equal to the value we think they're equal to. but logically we can think of it that way.
- But in real term they're just memory addresses.
